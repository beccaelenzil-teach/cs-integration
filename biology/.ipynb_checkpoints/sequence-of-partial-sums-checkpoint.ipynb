```python
import math
import matplotlib.pyplot as plt


def f1(n):
    return math.factorial(n-1)/n*n

def f2(n):
    return (math.e)**n

def f3(n):
    return math.tan(1/n)

def f4(n):  
    return (1/n**2)*math.tan(1/n)

def f5(n):
    return (-1)**(n-1)*2**n/(n**2)

def f6(n):
    return math.sin(n**2)/(n**2)
def f7(n):
    return (n+2)/(n**2-1)
    
def f8(n):
    return 1/n



def sequence(function, i):
    sum_list = []
    n_list = []
    the_sum = 0

    for n in range(1,i):
        try:
            the_sum += function(n)
        except:
            the_sum = the_sum
        sum_list.append(the_sum)
        n_list.append(n)
        
    plt.plot(n_list, sum_list)
    plt.show()
        
print('(n-1)!/n^2')
sequence(f1, 100)
print('e^2')
sequence(f2, 100)
print('tan(1/n)')
sequence(f3, 100)
print('(math.e)**n')
sequence(f4, 100)
print('(-1)**(n-1)*2**n/(n**2)')
sequence(f5, 100)
print('sin(n^2)/(n^2)')
sequence(f6, 100)
print('(n+2)/(n^2-1)')
sequence(f7, 100)
sequence(f8, 100)

        
    
    

        
```
