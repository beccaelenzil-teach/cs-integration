```python
import random
import matplotlib.pyplot as plt
import time

def genotype_frequency_simulation(fAA, faa, fAa, n, generations):
    fA = fAA + fAa/2
    fa = faa + fAa/2

    population = []
    fA_list = []
    fa_list = []
    fAA_list = []
    fAa_list = []
    faa_list = []
    gen_list = []
    
    
    for gen in range(generations):
        na = 0
        nA = 0
        
        nAA = 0
        nAa = 0
        naa = 0

        for i in range(n):
            organism = ''
            for k in range(2):
                seed = random.random()
                if seed < fA:
                    organism += 'A'
                    nA += 1
                else:
                    organism += 'a'
                    na += 1

            if organism == 'AA':
                nAA += 1
            elif organism == 'Aa' or organism == 'aA':
                nAa += 1
            elif organism == 'aa':
                naa += 1

            population.append(organism)


        fA = na/(2*n)
        fa = na/(2*n)
        
        fA_list.append(fA)
        fa_list.append(fa)
        fAA_list.append(fAA)
        fAa_list.append(fAa)
        faa_list.append(faa)
        gen_list.append(gen)

        fAA = nAA/n
        fAa = nAa/n
        faa = naa/n 
    
    
    plt.plot(gen_list, fAA_list, 'r', gen_list, fAa_list, 'b', gen_list, faa_list,'g', lineWidth = 2.0)
    plt.axis([0, generations, 0, 1])
    plt.ylabel('Genotype Frequencies')
    plt.show()
    time.sleep(1)
    
    
def hardy_weinberg(fAA, faa, fAa):
    
    fA_list = []
    fa_list = []
    fAA_list = []
    fAa_list = []
    faa_list = []
    i_list = []
    
    for i in range(10):
        fA = fAA + fAa/2
        fa = faa + fAa/2
        
        fA_list.append(fA)
        fa_list.append(fa)
        fAA_list.append(fAA)
        fAa_list.append(fAa)
        faa_list.append(faa)
        i_list.append(i)

        fAA = fA**2
        fAa = 2*fA*fa
        faa = fa**2
    
    #print("Genotype Frequency(AA) = ", fAA_list)
    #print("Genotype Frequency(Aa) = ", fAa_list)
    #print("Genotype Frequency(aa) = ", faa_list)
    
    
    plt.plot(i_list, fAA_list, 'r', i_list, fAa_list, 'b', i_list, faa_list,'g', lineWidth = 2.0)
    plt.ylabel('Genotype Frequencies')
    plt.axis([0,10,0,1])
    plt.title("Hardy Weinberg")
    plt.show()

    

for n in [4, 16, 80, 100, 1000, 10000, 50000]:
    print("Population = ", n)
    genotype_frequency_simulation(.2, .2, .6, n, 20)

hardy_weinberg(.2, .2, .6)
        
    
```
